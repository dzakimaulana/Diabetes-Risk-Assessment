## Dataset Explanation





**Dataset Descriptions**

The dataset used is a clean dataset of 253,680 survey responses to CDC'S BRFSS2015. It has a target variable called "Diabetes_012" that contain 3 classes. This variable indicates wheter or not the person have diabetes. 0 = no diabetes, 1 = prediabetes and 2 = diabetes. This dataset used 21 variables to determine wheter or not a person have diabetes.

**Notes:**

  **Majority of variables in the dataset used 0 as no and 1 as yes, except for these variables: BMI, GenHlth, MentHlth, PhysHlth, Age, Education, and Income**
1. **HighBP**

  Indicates High Blood Pressure.
2. **HighChol**

   Indicates high Cholesterol.
3. **CholCheck**

   Indicates cholesterol check in the interval of 5 years.
4. **BMI**

  Indicates the Body Mass Index of the person.
5. **Smoker**

  Indicates smoked at least 100 cigarettes (5 Packs) in their life.
6. **Stroke**

   Indicates have been told they have stroke.
7. **HeartDiseaseorAttack**

  Indicates have a heart disease or attack (Coronary Heart Disease [CHD] and Myocardial Infarction).
8. **PhysActivity**

  Indicates physical activity in the past 30 days (not including job).
9. **Fruits**

  Indicates consuming fruit 1 or more time per day.
10. **Veggies**
  
  Indicates consuming vegetables 1 or more time per day.
11. **HvyAkcoholConsump**

  Indicates a heavy drinkers (For adult men having more than 14 drinks per week and adult woman having more than 7 drinks per week).
12. **AnyHealtcare**

  Indicates having any kind of health care coverage.
13. **NoDocbcCost**
  
  Indicates past instances where financial constraints prevented doctor visits.
14. **GenHlth**

  Indicates overall health they think they have from the scale of 1-5
15. **MentHlth**

  Indicates how many days, the state of mental health where they were not good in the past 30 days.
16. **PhyshLTH**

  Indicates how many days, the state of physical health where they were not good in the past 30 days.
17. **DiffWalk**

  Indicates having serious difficulty walking or climbing staits.
18. **Sex**

  Indicates gender (0 = female and 1 = male).
19. **Age**

  Indicates age category (1=18-24, 9=60-64, 13=80 or older).
20. **Education**

  Indicates Education level (1 = Never attended school or only kindergarten, 2 = Grades 1 through 8 (Elementary), 3 = Grades 9 through 11 (Some high school), 4 = Grade 12 or GED (High school graduate), 5 = College 1 year to 3 years (Some college or technical school), and 6 = College 4 years or more (College)).
21. **Income**

  Indicates the income from the scale of 1-8 (1 = less than $10,000, 5 = less than $35,000, and 8 = $75,000 or more)

## Data Preprocessing

```python
data.value_counts('Diabetes_012')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer, MinMaxScaler
from sklearn.preprocessing import StandardScaler
from keras.utils import to_categorical
from imblearn.over_sampling import SMOTE
```

```python
X = data.drop(columns=['Diabetes_012'], axis=1)
y = data['Diabetes_012']
```

```python
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_test
```

```python
y_train_categories = to_categorical(y_train)
y_test_categories = to_categorical(y_test)
y_test_categories
```

## Model Training



```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Input, Dropout
```

```python
model = Sequential([
    Input(shape=(21,)),
    Dense(units=32, activation='relu'),
    Dropout(0.2),
    Dense(units=16, activation='relu'),
    Dropout(0.2),
    Dense(units=3, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
print("[INFO] training network...")
H = model.fit(X_train, y_train_categories, validation_data=(X_test, y_test_categories), epochs=10, batch_size=128)
```

## Model Evaluation

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score, roc_curve
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test_categories)
print(f'Loss: {test_loss}')
print(f'Accuracy: {test_accuracy}')
```

```python
import matplotlib.pyplot as plt

# Plot training & validation loss values
plt.plot(H.history['loss'])
plt.plot(H.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

# Plot training & validation accuracy values
plt.plot(H.history['accuracy'])
plt.plot(H.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
# Predict the values from the validation dataset
y_pred = model.predict(X_test)
# Convert predictions classes to one hot vectors
y_pred_classes = np.argmax(y_pred,axis = 1)
# Convert validation observations to one hot vectors
y_true = np.argmax(y_test_categories,axis = 1)
# compute the confusion matrix
confusion_mtx = confusion_matrix(y_true, y_pred_classes)
# plot the confusion matrix
f,ax = plt.subplots(figsize=(9, 9))
sns.heatmap(confusion_mtx, annot=True, linewidths=0.01,cmap="Greens",linecolor="gray", fmt= '.1f',ax=ax)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()
```
